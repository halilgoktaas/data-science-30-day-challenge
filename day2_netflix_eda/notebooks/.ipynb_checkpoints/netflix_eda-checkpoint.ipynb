```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.style.use("ggplot")
```

```python
df = pd.read_csv("netflix_titles.csv", encoding = "latin1")
```

```python
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
```

```python
df = df.drop(columns = ["."], errors='ignore')
```

```python
df.shape
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
unnamed_cols = [c for c in df.columns if c.lower().startswith("unnamed")]
```

```python
df = df.drop(columns=unnamed_cols)
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors = "coerce")
```

```python
df["year_added"] = df["date_added"].dt.year
```

```python
df['year_added'] = df['date_added'].dt.year.astype("Int64")
```

```python
df["year_added"]
```

```python
df["type"].value_counts()
```

```python
df['duration_num'] = df['duration'].str.extract(r"(\d+)").astype('float')
```

```python
df['duration_unit'] = df['duration'].str.extract(r"([A-Za-z]+)")
```

```python
df['minutes'] = df.apply(lambda x: x['duration_num'] if x['type'] == 'Movie' else None, axis = 1 )
```

```python
df['seasons'] = df.apply(lambda x: x['duration_num'] if x['type'] == 'TV Show' else None, axis = 1 )
```

```python
genres = (df['listed_in'].str.split(",").explode().value_counts().head(10))
```

```python
genres
```

```python
countries = (df['country'].fillna("Unknown").str.split(",").explode().value_counts().head(10))
```

```python
countries
```

```python
sns.countplot(data=df, x='type')
plt.title('movies vs tv shows')
plt.show()
```

```python
df["year_added"].value_counts().sort_index().plot(figsize = (10,5))
plt.title("content added over years")
plt.xlabel("year")
plt.ylabel("count")
plt.show()
```

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("top 10 content producing countries")
plt.show()
```

```python
df["rating"].value_counts().head(10).plot(kind="bar")
```

## Key Insights

- Netflix has significantly increased content production after 2016.
- Movies dominate the platform compared to TV shows.
- The United States is the leading content producer.
- Most content is targeted toward mature audiences.
