| Segment ID | Segment Adı |
|-----------|-------------|
| 2 | Champions |
| 1 | Loyal Customers |
| 0 | Potential / Regular Customers |
| 3 | Lost / At Risk Customers |

---

## Segment Profilleri ve İş Yorumları

### Champions (Segment 2)

Bu segment, son dönemde alışveriş yapan, yüksek alışveriş sıklığına sahip ve en yüksek toplam harcamayı gerçekleştiren müşteri grubudur. Müşteri sayısı görece az olmasına rağmen toplam gelire katkısı oldukça yüksektir.

**İş Yorumu:**  
Bu müşteriler işletmenin en değerli müşteri grubunu oluşturmaktadır. Fiyat odaklı kampanyalar yerine özel hizmetler, sadakat programları ve kişiselleştirilmiş deneyimler tercih edilmelidir.

---

### Loyal Customers (Segment 1)

Bu segment düzenli alışveriş yapan ve orta-yüksek seviyede harcama gerçekleştiren müşterilerden oluşmaktadır. Champions segmentine dönüşme potansiyeline sahiptir.

**İş Yorumu:**  
Sadakat programları, çapraz satış ve kişiselleştirilmiş teklifler ile bu segmentin müşteri yaşam boyu değeri artırılabilir.

---

### Potential / Regular Customers (Segment 0)

Bu segment en kalabalık müşteri grubunu temsil etmektedir. Alışveriş sıklığı ve harcama miktarı düşük olmasına rağmen büyüme potansiyeli yüksektir.

**İş Yorumu:**  
Hedefli kampanyalar, indirim stratejileri ve doğru iletişim kanalları kullanılarak bu segmentten elde edilen gelir artırılabilir.

---

### Lost / At Risk Customers (Segment 3)

Bu segment uzun süredir alışveriş yapmamış, düşük alışveriş sıklığına ve düşük harcama değerlerine sahip müşterilerden oluşmaktadır.

**İş Yorumu:**  
Bu grup için geniş çaplı pazarlama harcamaları verimsiz olabilir. Seçici ve düşük maliyetli reaktivasyon stratejileri uygulanmalıdır.

---

## Genel İş İçgörüleri

- Küçük bir müşteri grubu toplam gelirin büyük bir bölümünü oluşturmaktadır.
- Sadık müşteriler doğru stratejilerle yüksek değerli müşteri segmentine dönüştürülebilir.
- En kalabalık müşteri segmenti, hedefli pazarlama çalışmaları ile önemli bir gelir artışı potansiyeline sahiptir.
- Uzun süredir pasif olan müşteriler için agresif pazarlama yerine seçici yaklaşımlar tercih edilmelidir.
