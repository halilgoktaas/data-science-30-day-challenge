## Segment Naming

The customer segments obtained using the KMeans algorithm were renamed to better reflect business understanding and decision-making needs.

| Segment ID | Segment Name |
|-----------|--------------|
| 2 | Champions |
| 1 | Loyal Customers |
| 0 | Potential / Regular Customers |
| 3 | Lost / At Risk Customers |

---

## Segment Profiles and Business Interpretation

### Champions (Segment 2)

This segment includes customers who purchased recently, shop frequently, and generate the highest total revenue. Although relatively small in size, this group contributes disproportionately to overall revenue.

**Business Interpretation:**  
These customers represent the most valuable segment. Loyalty-focused strategies and personalized experiences should be prioritized over price-based discounts.

---

### Loyal Customers (Segment 1)

Customers in this segment make regular purchases and generate moderate to high revenue. They have strong potential to be converted into Champions.

**Business Interpretation:**  
Loyalty programs, cross-selling, and personalized offers can significantly increase customer lifetime value.

---

### Potential / Regular Customers (Segment 0)

This is the largest customer segment, characterized by moderate engagement and relatively low spending per customer.

**Business Interpretation:**  
Targeted promotions and marketing campaigns can unlock substantial revenue growth from this group.

---

### Lost / At Risk Customers (Segment 3)

This segment consists of customers who have not made purchases for a long time and exhibit low engagement and spending levels.

**Business Interpretation:**  
Selective reactivation campaigns are recommended instead of broad marketing efforts to optimize budget efficiency.

---

## Key Business Insights

- A small fraction of customers generates a significant portion of total revenue.
- Loyal customers represent a strong opportunity to grow high-value customer segments.
- The largest customer segment holds considerable revenue growth potential when properly targeted.
- Selective reactivation strategies are more efficient than aggressive marketing for inactive customers.
